```python
import urllib.request
import json
import re

req = urllib.request.Request(
    'https://www.filmhouseng.com/en/cinemas/lekki/movies', 
    headers={'User-Agent': 'Mozilla/5.0'}
)
html = urllib.request.urlopen(req).read().decode('utf-8')
m = re.search(r'<script id="__NEXT_DATA__"[^>]*>([\s\S]*?)</script>', html)
if m:
    data = json.loads(m.group(1))
    def walk(obj, path=""):
        if isinstance(obj, dict):
            for k, v in obj.items():
                if k in ['props', 'pageProps', 'movies', 'showtimes', 'cinema', 'data']:
                    print(f"{path}.{k} (dict)")
                    walk(v, path + "." + k)
        elif isinstance(obj, list) and len(obj) > 0 and isinstance(obj[0], dict):
            print(f"{path} -> list of {len(obj)} items. Keys of first: {list(obj[0].keys())}")
            if 'title' in obj[0]:
                print(json.dumps(obj[0], indent=2)[:1000])
    walk(data)
else:
    print("__NEXT_DATA__ not found")
```
